# Probability Density Function


### Explanation


A probability density function (pdf) is used to represent a continuous probability distribution. Unlike a discrete probability distribution, where each outcome has a specific probability given by the probability mass function (pmf), a continuous random variable takes on values over a continuum (often an interval). 


For a continuous variable:

- The probability of the variable taking on any exact value is zero.

- Instead, probabilities are determined by integrating the pdf over an interval.


Thus, the pdf gives us the density of the probability at every point, and the probability that the variable falls within an interval [a, b] is calculated as:

\[
P(a \leq X \leq b) = \int_a^b f(x) \, dx
\]

This is why a continuous probability distribution is represented by a Probability density function rather than a Probability mass function, which is used exclusively for discrete distributions

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def main():
    # Parameters for the normal distribution
    mu = 0       # Mean
    sigma = 1    # Standard Deviation

    # Create an array of x values from -5 to 5
    x = np.linspace(-5, 5, 1000)

    # Compute PDF values for the normal distribution at each x value
    pdf_values = norm.pdf(x, mu, sigma)

    # Print the first 5 x values and PDF values as a sample output
    print("First 5 x values:", x[:5])
    print("First 5 PDF values:", pdf_values[:5])

    # Plot the probability density function
    plt.figure(figsize=(8, 6))
    plt.plot(x, pdf_values, label=f'Normal PDF (μ={mu}, σ={sigma})')
    plt.title("Probability Density Function of a Normal Distribution")
    plt.xlabel("x")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    # -------------------------------
    # Generate dummy data from the normal distribution
    dummy_data = np.random.normal(mu, sigma, 1000)
    print("\nDummy data sample (first 10 values):", dummy_data[:10])

    # Plot the histogram of the dummy data (normalized to form a density)
    # and overlay the theoretical normal PDF
    plt.figure(figsize=(8, 6))
    plt.hist(dummy_data, bins=30, density=True, alpha=0.6, color='g', label='Dummy Data Histogram')
    plt.plot(x, pdf_values, 'r-', lw=2, label='Normal PDF')
    plt.title("Dummy Data Histogram with Normal PDF")
    plt.xlabel("x")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
````

````python
# example of a probability density function with exponential distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

def main():
    # Set the scale parameter for the exponential distribution.
    # The exponential distribution PDF is defined for x >= 0 as:
    # f(x) = (1/scale) * exp(-x/scale)
    scale = 1.0  # mean (and standard deviation for the exponential distribution)

    # Create an array of x values from 0 to 10.
    x = np.linspace(0, 10, 1000)

    # Compute the PDF values for the exponential distribution.
    pdf_values = expon.pdf(x, scale=scale)
    
    # Print the first 5 x values and their corresponding PDF values.
    print("First 5 x values:", x[:5])
    print("First 5 PDF values:", pdf_values[:5])
    
    # Plot the probability density function.
    plt.figure(figsize=(8, 6))
    plt.plot(x, pdf_values, label=f'Exponential PDF (scale={scale})', color='blue')
    plt.title("Probability Density Function of an Exponential Distribution")
    plt.xlabel("x")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    # -------------------------------
    # Generate dummy data from the exponential distribution.
    dummy_data = np.random.exponential(scale, 1000)
    print("\nDummy data sample (first 10 values):", dummy_data[:10])
    
    # Plot the histogram of the dummy data (normalized as a density)
    # and overlay the theoretical exponential PDF.
    plt.figure(figsize=(8, 6))
    plt.hist(dummy_data, bins=30, density=True, alpha=0.6, color='green', label='Dummy Data Histogram')
    plt.plot(x, pdf_values, 'r-', lw=2, label='Exponential PDF')
    plt.title("Histogram of Dummy Data with Exponential PDF")
    plt.xlabel("x")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
````

````python
# python code for the binomial distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def main():
    # Parameters for the Binomial distribution
    n = 10    # number of trials
    p = 0.5   # probability of success

    # Generate the range of outcomes: 0 through n
    x = np.arange(0, n + 1)  # possible number of successes

    # Calculate the theoretical PMF for each outcome
    pmf_values = binom.pmf(x, n, p)
    
    # Print the outcomes and PMF values
    print("Possible outcomes (number of successes):", x)
    print("PMF values:", pmf_values)

    # Generate dummy data by sampling from the Binomial distribution
    dummy_data = np.random.binomial(n, p, size=1000)
    print("\nDummy data sample (first 10 values):", dummy_data[:10])
    
    # Plot the theoretical PMF
    plt.figure(figsize=(8, 6))
    plt.plot(x, pmf_values, 'bo', ms=8, label='Theoretical Binomial PMF')
    plt.vlines(x, 0, pmf_values, colors='b', lw=2, alpha=0.5)
    
    # Plot a normalized histogram of the dummy data
    plt.hist(dummy_data, bins=np.arange(-0.5, n+1.5, 1), density=True,
             alpha=0.6, color='orange', label='Dummy Data Histogram')
    
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.title(f'Binomial Distribution PMF (n={n}, p={p})')
    plt.legend()
    plt.grid(True)
    plt.show()
    
if __name__ == "__main__":
    main()
````

## A NORMAL Distribution FACTS

- A normal distribution is a Continuous Distribution

- In a normal distribution , data percentages are distributed across 1, 2 and 3 standard deviations as 68, 95% and 99.7% of the data respectively. This is known as the 68-98-99.7 rule or the empirical rule in statistics

- A standard normal distribution has a mean equal to 0 and a standard deviation of 1 

- In a normal distribution data above 1 standard deviation away and less than 2 stddev is of 27%

- Problem example: 

The mean weight of a morning breakfast cereal pack is 0.295 kg with a standard deviation of 0.025 kg. The random variable weight of the pack follows a normal distribution. How many standard deviations away from the mean is a pack that weighs 0.260 kg?

- Problem example:

- Process of stardadization -> (X - mu)/sigma



### Explanation

A standard normal distribution (also known as a z-distribution) is a special case of the normal distribution where:
- Mean (μ) = 0
- Standard Deviation (σ) = 1



### Key Properties of Standard Normal Distribution

1. **Mean (μ) = 0**
   - The center of the distribution is at zero
   - The distribution is symmetric around zero

2. **Standard Deviation (σ) = 1**
   - One standard deviation equals one unit on the x-axis
   - Makes it easier to calculate probabilities

3. **Other Properties:**
   - Variance (σ²) = 1 (since variance is the square of standard deviation)
   - Symmetric bell-shaped curve
   - Total area under the curve = 1

### Why the Other Options are Wrong:

1. "The mean is equal to the variance"
   - In standard normal, mean = 0, variance = 1
   - Therefore, they are not equal

2. "The mean and standard deviation are both equal to 1"
   - Mean = 0, not 1
   - Only the standard deviation equals 1

3. "The mean is equal to the standard deviation"
   - Mean = 0
   - Standard deviation = 1
   - Therefore, they are not equal

The standard normal distribution is particularly useful because:
- Any normal distribution can be converted to standard normal using z-scores
- It simplifies probability calculations
- It's widely used in statistical inference and hypothesis testing


````python
# Normal Distribution in Python Example of Continuous Distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create points for x-axis
x = np.linspace(-4, 4, 1000)

# Calculate PDF for standard normal distribution (mean=0, std=1)
pdf = norm.pdf(x, loc=0, scale=1)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, 'b-', label='Normal Distribution')
plt.fill_between(x, pdf, alpha=0.3)
plt.title('Normal Distribution (μ=0, σ=1)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True, alpha=0.3)
plt.legend()

# Add vertical lines for standard deviations
for i in [-3, -2, -1, 0, 1, 2, 3]:
    plt.axvline(x=i, color='r', linestyle='--', alpha=0.2)

plt.show()
````

````python
# empirical rule in python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create points for x-axis
x = np.linspace(-4, 4, 1000)

# Calculate PDF for standard normal distribution
pdf = norm.pdf(x, loc=0, scale=1)

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(x, pdf, 'b-', label='Normal Distribution')

# Fill areas for different standard deviations
colors = ['red', 'green', 'blue']
alphas = [0.1, 0.1, 0.1]
labels = ['68% (±1σ)', '95% (±2σ)', '99.7% (±3σ)']
std_ranges = [(-1, 1), (-2, 2), (-3, 3)]

for (start, end), color, alpha, label in zip(std_ranges, colors, alphas, labels):
    mask = (x >= start) & (x <= end)
    plt.fill_between(x[mask], pdf[mask], color=color, alpha=alpha, label=label)

plt.title('Normal Distribution with Standard Deviations')
plt.xlabel('Standard Deviations from Mean')
plt.ylabel('Probability Density')
plt.grid(True, alpha=0.3)
plt.legend()

# Add vertical lines for standard deviations
for i in [-3, -2, -1, 0, 1, 2, 3]:
    plt.axvline(x=i, color='gray', linestyle='--', alpha=0.5)

plt.show()

# Print the exact probabilities
print(f"Probability within 1 standard deviation: {norm.cdf(1) - norm.cdf(-1):.1%}")
print(f"Probability within 2 standard deviations: {norm.cdf(2) - norm.cdf(-2):.1%}")
print(f"Probability within 3 standard deviations: {norm.cdf(3) - norm.cdf(-3):.1%}")
````

````python
# Standard Normal Distribution Deviation above 1 and below 2

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create points for x-axis
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, loc=0, scale=1)

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(x, pdf, 'b-', label='Normal Distribution')

# Shade the area between 1 and 2 standard deviations (both sides)
mask_right = (x >= 1) & (x <= 2)
mask_left = (x >= -2) & (x <= -1)

plt.fill_between(x[mask_right], pdf[mask_right], color='red', alpha=0.3, 
                 label='Between 1σ and 2σ (13.5% each side)')
plt.fill_between(x[mask_left], pdf[mask_left], color='red', alpha=0.3)

plt.title('Normal Distribution: Area Between 1 and 2 Standard Deviations')
plt.xlabel('Standard Deviations from Mean')
plt.ylabel('Probability Density')
plt.grid(True, alpha=0.3)
plt.legend()

# Add vertical lines for standard deviations
for i in [-2, -1, 0, 1, 2]:
    plt.axvline(x=i, color='gray', linestyle='--', alpha=0.5)

plt.show()

# Calculate exact probabilities
within_2sd = norm.cdf(2) - norm.cdf(-2)  # 95%
within_1sd = norm.cdf(1) - norm.cdf(-1)  # 68%
between_1_2sd = within_2sd - within_1sd   # 27%

print(f"Percentage between 1 and 2 standard deviations: {between_1_2sd:.1%}")
print(f"(This is split as {between_1_2sd/2:.1%} on each side)")
````

````python
# cereal pack example

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given values
mean = 0.295  # kg
std_dev = 0.025  # kg
x = 0.260  # kg

# Calculate z-score
z_score = (x - mean) / std_dev
print(f"Z-score: {z_score}")

# Create points for plotting
weights = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
pdf = norm.pdf(weights, mean, std_dev)

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(weights, pdf, 'b-', label='Normal Distribution')

# Mark the specific weight
plt.axvline(x=x, color='r', linestyle='--', 
            label=f'Weight: {x}kg\n(z = {z_score:.1f}σ)')
plt.axvline(x=mean, color='g', linestyle='--', 
            label=f'Mean: {mean}kg')

plt.title('Normal Distribution of Cereal Pack Weights')
plt.xlabel('Weight (kg)')
plt.ylabel('Probability Density')
plt.grid(True, alpha=0.3)
plt.legend()

# Add annotation
plt.annotate(f'{z_score:.1f}σ',
            xy=(x, 0),
            xytext=(x, max(pdf)/2),
            ha='right',
            va='center',
            arrowprops=dict(arrowstyle='->'))

plt.show()
````

# delivery time example 

The mean delivery time for a food delivery company is normally distributed with a mean of 47.5 minutes and a standard deviation of 8.7 minutes.

68% of the deliveries will be done between _____ minutes and ___ minutes.

````python
# delivery time example

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mean = 47.5
std_dev = 8.7

# Create points for x-axis (delivery times)
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
pdf = norm.pdf(x, mean, std_dev)

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(x, pdf, 'b-', label='Normal Distribution')

# Shade the area within 1 standard deviation (68%)
mask = (x >= mean-std_dev) & (x <= mean+std_dev)
plt.fill_between(x[mask], pdf[mask], color='red', alpha=0.3, 
                 label='68% of deliveries\n(38.8 to 56.2 minutes)')

# Add vertical lines for mean and ±1 standard deviation
plt.axvline(x=mean, color='green', linestyle='--', label=f'Mean: {mean} minutes')
plt.axvline(x=mean-std_dev, color='red', linestyle='--', 
            label=f'-1σ: {mean-std_dev:.1f} minutes')
plt.axvline(x=mean+std_dev, color='red', linestyle='--', 
            label=f'+1σ: {mean+std_dev:.1f} minutes')

plt.title('Delivery Time Distribution')
plt.xlabel('Delivery Time (minutes)')
plt.ylabel('Probability Density')
plt.grid(True, alpha=0.3)
plt.legend()

# Print the exact values
print(f"68% of deliveries will be between:")
print(f"Lower bound (μ - σ): {mean-std_dev:.1f} minutes")
print(f"Upper bound (μ + σ): {mean+std_dev:.1f} minutes")

plt.show()
````

# normal distribution hands on

- CDF 

CDF stands for Cumulative Distribution Function. It calculates the probability that a random variable X takes on a value less than or equal to a specific value

- PDF 
Probability Density Function


- To calculate the proability of a certain value bellow a certain value, we use CDF and PDF and sigma or std deviation




````python
import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('sat_score.csv', header=0)

# Calculate mean and standard deviation
mean = data['score'].mean()
std = data['score'].std()

# Calculate probability for score <= 750
prob = norm.cdf(750, mean, std)

# Visualization
x = np.linspace(mean - 4*std, mean + 4*std, 1000)
pdf = norm.pdf(x, mean, std)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, 'b-', label='Normal Distribution')

# Shade area for scores <= 750
mask = x <= 750
plt.fill_between(x[mask], pdf[mask], color='red', alpha=0.3, 
                 label=f'P(X ≤ 750) = {prob:.3f}')

plt.axvline(x=750, color='red', linestyle='--', label='X = 750')
plt.axvline(x=mean, color='green', linestyle='--', label=f'Mean = {mean:.1f}')

plt.title('SAT Score Distribution')
plt.xlabel('SAT Score')
plt.ylabel('Probability Density')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

print(f"\nMean SAT score: {mean:.2f}")
print(f"Standard deviation: {std:.2f}")
print(f"Probability of scoring <= 750: {prob:.3f}")
````

## If n(>30) samples are drawn from a population that has a mean μ and standard deviation σ. The sampling distribution will follow a normal distribution with:

Mean: μ; Standard Deviation: σ / √n
This is known as the Central Limit Theorem (CLT).

````python

# Central Limit Theorem Demonstration
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def demonstrate_sampling_distribution(population_mean=100, population_std=15, sample_size=30, n_samples=1000):
    # Create a non-normal population (e.g., exponential distribution shifted by population_mean)
    population = np.random.exponential(scale=population_std, size=10000) + population_mean
    
    # Take multiple samples and calculate their means
    sample_means = []
    for _ in range(n_samples):
        sample = np.random.choice(population, size=sample_size)
        sample_means.append(np.mean(sample))
    
    # Calculate theoretical sampling distribution parameters
    sampling_mean = population_mean
    sampling_std = population_std / np.sqrt(sample_size)
    
    # Create plots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Plot population distribution
    ax1.hist(population, bins=50, density=True, alpha=0.7, color='blue',
             label='Population Distribution')
    ax1.axvline(population_mean, color='red', linestyle='--',
                label=f'Population Mean: {population_mean}')
    ax1.set_title('Population Distribution')
    ax1.legend()
    
    # Plot sampling distribution
    ax2.hist(sample_means, bins=50, density=True, alpha=0.7, color='green',
             label='Sampling Distribution')
    
    # Add theoretical normal curve
    x = np.linspace(min(sample_means), max(sample_means), 100)
    y = norm.pdf(x, sampling_mean, sampling_std)
    ax2.plot(x, y, 'r-', label='Theoretical Normal')
    ax2.axvline(np.mean(sample_means), color='red', linestyle='--',
                label=f'Sample Mean: {np.mean(sample_means):.1f}')
    ax2.set_title(f'Sampling Distribution (n={sample_size})')
    ax2.legend()
    
    plt.suptitle('Central Limit Theorem Demonstration')
    plt.show()
    
    # Print results
    print(f"Population Parameters:")
    print(f"Mean: {population_mean}")
    print(f"Standard Deviation: {population_std}")
    print(f"\nSampling Distribution Parameters:")
    print(f"Theoretical Mean: {sampling_mean}")
    print(f"Theoretical Std: {sampling_std:.2f}")
    print(f"Observed Mean: {np.mean(sample_means):.2f}")
    print(f"Observed Std: {np.std(sample_means):.2f}")

# Demonstrate
demonstrate_sampling_distribution()
````

````python
# standard error positive or negative?

# The standard error is always positive. It measures the variability of the sample mean around the population mean.

import numpy as np
import matplotlib.pyplot as plt

def calculate_standard_error(std_dev, sample_sizes):
    """Calculate standard error for different sample sizes"""
    return std_dev / np.sqrt(sample_sizes)

# Create range of sample sizes
sample_sizes = np.arange(10, 1000, 10)
population_std = 15  # example standard deviation

# Calculate standard errors
standard_errors = calculate_standard_error(population_std, sample_sizes)

# Visualization
plt.figure(figsize=(12, 6))
plt.plot(sample_sizes, standard_errors, 'b-', label='Standard Error')
plt.fill_between(sample_sizes, standard_errors, alpha=0.3)

plt.title('Standard Error vs Sample Size')
plt.xlabel('Sample Size (n)')
plt.ylabel('Standard Error')
plt.grid(True, alpha=0.3)
plt.legend()

# Add horizontal line at y=0 to show SE never goes negative
plt.axhline(y=0, color='r', linestyle='--', alpha=0.5, 
            label='Zero Line (SE never goes below)')

plt.show()

# Print some example values
print("Example Standard Errors for different sample sizes:")
for n in [10, 30, 100, 500]:
    se = population_std / np.sqrt(n)
    print(f"n = {n:4d}: SE = {se:.4f}")


````

````python
# demonstration of statemnents about CLT

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def demonstrate_clt_assumptions(n_samples=1000):
    fig, axes = plt.subplots(3, 2, figsize=(15, 12))
    
    # 1. Random Sampling Demonstration
    # Compare random vs non-random sampling
    population = np.random.exponential(size=10000)
    
    # Random sampling
    random_samples = [np.mean(np.random.choice(population, 30)) 
                     for _ in range(n_samples)]
    
    # Non-random sampling (taking first 30 elements repeatedly)
    nonrandom_samples = [np.mean(population[:30]) 
                        for _ in range(n_samples)]
    
    axes[0,0].hist(random_samples, bins=30, density=True, alpha=0.7,
                   label='Random Sampling')
    axes[0,0].set_title('Random Sampling\n(Follows CLT)')
    axes[0,1].hist(nonrandom_samples, bins=30, density=True, alpha=0.7,
                   label='Non-random Sampling')
    axes[0,1].set_title('Non-random Sampling\n(Violates CLT)')
    
    # 2. Independence Demonstration
    # Compare independent vs dependent sampling
    independent_samples = [np.mean(np.random.choice(population, 30)) 
                         for _ in range(n_samples)]
    
    # Dependent samples (moving average)
    dependent_samples = np.convolve(population, 
                                  np.ones(30)/30, 
                                  mode='valid')[:n_samples]
    
    axes[1,0].hist(independent_samples, bins=30, density=True, alpha=0.7,
                   label='Independent Samples')
    axes[1,0].set_title('Independent Samples\n(Follows CLT)')
    axes[1,1].hist(dependent_samples, bins=30, density=True, alpha=0.7,
                   label='Dependent Samples')
    axes[1,1].set_title('Dependent Samples\n(Violates CLT)')
    
    # 3. Sample Size Demonstration
    small_samples = [np.mean(np.random.choice(population, 5)) 
                    for _ in range(n_samples)]
    large_samples = [np.mean(np.random.choice(population, 30)) 
                    for _ in range(n_samples)]
    
    axes[2,0].hist(small_samples, bins=30, density=True, alpha=0.7,
                   label='Small Sample (n=5)')
    axes[2,0].set_title('Small Sample Size\n(May Not Follow CLT)')
    axes[2,1].hist(large_samples, bins=30, density=True, alpha=0.7,
                   label='Large Sample (n=30)')
    axes[2,1].set_title('Large Sample Size\n(Follows CLT)')
    
    plt.tight_layout()
    plt.suptitle('Demonstration of CLT Assumptions', y=1.02)
    plt.show()

# Run demonstration
demonstrate_clt_assumptions()

# Print key points about each assumption
print("\nCLT Assumptions:")
print("\n1. Random Sampling:")
print("- Ensures unbiased representation of population")
print("- Eliminates systematic errors")
print("- Allows for valid statistical inference")

print("\n2. Independence:")
print("- Each sample value must not influence others")
print("- No systematic relationships between observations")
print("- Essential for valid probability calculations")

print("\n3. Sample Size:")
print("- Generally n ≥ 30 is considered sufficient")
print("- Larger samples needed for skewed distributions")
print("- Affects convergence to normal distribution")
````

````python
# point estimate variability in sampling distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def demonstrate_point_estimation(population_mean=100, population_std=15, 
                               sample_size=30, n_samples=1000):
    """
    Demonstrate point estimation and its properties
    """
    # Create population
    population = np.random.normal(population_mean, population_std, 10000)
    
    # Take multiple samples and calculate point estimates
    sample_means = []
    for _ in range(n_samples):
        sample = np.random.choice(population, size=sample_size)
        sample_means.append(np.mean(sample))
    
    # Create visualization
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Plot population distribution
    ax1.hist(population, bins=50, density=True, alpha=0.7, color='blue',
             label='Population')
    ax1.axvline(population_mean, color='red', linestyle='--',
                label=f'Population Mean: {population_mean}')
    ax1.set_title('Population Distribution')
    ax1.legend()
    
    # Plot sampling distribution of means
    ax2.hist(sample_means, bins=50, density=True, alpha=0.7, color='green',
             label='Sample Means')
    ax2.axvline(np.mean(sample_means), color='red', linestyle='--',
                label=f'Mean of Means: {np.mean(sample_means):.1f}')
    ax2.axvline(population_mean, color='blue', linestyle='--',
                label=f'Population Mean: {population_mean}')
    ax2.set_title(f'Sampling Distribution of Means\n(n={sample_size})')
    ax2.legend()
    
    plt.suptitle('Point Estimation Demonstration')
    plt.show()
    
    # Print statistics
    print("\nPoint Estimation Statistics:")
    print(f"Population Mean: {population_mean}")
    print(f"Average of Sample Means: {np.mean(sample_means):.2f}")
    print(f"Standard Error: {np.std(sample_means):.2f}")
    print(f"95% of estimates fall between: "
          f"{np.percentile(sample_means, 2.5):.2f} and "
          f"{np.percentile(sample_means, 97.5):.2f}")
    
    # Calculate properties
    bias = np.mean(sample_means) - population_mean
    mse = np.mean((np.array(sample_means) - population_mean)**2)
    
    print(f"\nEstimator Properties:")
    print(f"Bias: {bias:.3f}")
    print(f"Mean Squared Error: {mse:.3f}")

# Run demonstration
demonstrate_point_estimation()
````

````python
# point estimate is a single value calculated from sample data that is used to estimate a population parameter.


# the sample mean is a point estimate of the population mean.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def demonstrate_point_estimation():
    # Create a population
    population_size = 10000
    population_mean = 100  # μ (population parameter)
    population_std = 15
    population = np.random.normal(population_mean, population_std, population_size)
    
    # Take a sample and calculate point estimate
    sample_size = 30
    sample = np.random.choice(population, sample_size)
    sample_mean = np.mean(sample)  # x̄ (point estimate)
    
    # Visualization
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Population distribution
    ax1.hist(population, bins=50, density=True, alpha=0.7, color='blue',
             label='Population')
    ax1.axvline(population_mean, color='red', linestyle='--',
                label=f'Population Mean (μ): {population_mean}')
    ax1.set_title('Population Distribution')
    ax1.legend()
    
    # Sample distribution
    ax2.hist(sample, bins=15, density=True, alpha=0.7, color='green',
             label='Sample')
    ax2.axvline(sample_mean, color='red', linestyle='--',
                label=f'Sample Mean (x̄): {sample_mean:.1f}')
    ax2.axvline(population_mean, color='blue', linestyle='--',
                label=f'Population Mean (μ): {population_mean}')
    ax2.set_title('Sample Distribution and Point Estimate')
    ax2.legend()
    
    plt.suptitle('Point Estimation: Sample Mean as an Estimate of Population Mean')
    plt.show()
    
    # Print results
    print("\nPoint Estimation Results:")
    print(f"Population Parameter (μ): {population_mean}")
    print(f"Point Estimate (x̄): {sample_mean:.2f}")
    print(f"Difference: {abs(population_mean - sample_mean):.2f}")
    
    return population, sample, population_mean, sample_mean

# Run demonstration
population, sample, pop_mean, sample_mean = demonstrate_point_estimation()
````
